```python
#Here we use anova/F test because there are 4 samples.when we want compare more than 2 samples we can use anova testing.
#It check the impact of one or more factors by comparing the means of diffrent sample
```

Alternate hypothesis(Ha): Atleast one the sample TAT population mean is different
