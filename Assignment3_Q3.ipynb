Q3:Sales of products in four diffrent regions is tabulated for males and females.Find if male-female buyer rations are similar across regions

```python
#Here we use chi-square Test
```

Chi-square test: Chi-square tests of independence test whether two qualitative variables are independent.That is whether there exist a relationship between two categorical variables.

Null hypothesis:The variables are independent,there is no dependency between the gender based buyer rations across regions

ie,Ho: The male-female buyer rations are similar across regions

Alternate hypothesis:The variables are dependent,there is significant dependency between the gender based buyer rations across regions.

ie,Ha:The male-female buyer rations are not similar across regions

```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
df=pd.read_csv("BuyerRatio.csv")
```

```python
df.head()
```

```python
df_table=df.iloc[:,1:6]
df_table
```

```python
df_table.values
```

```python
value=stats.chi2_contingency(df_table)
value
```

```python
#degrees of freedom=(no. of rows-1)*(no. of columns-1)
dof=value[2]
dof
```

```python
Expected_value=value[3]
Expected_value
```

```python
from scipy.stats import chi2
chi_square=sum([(o-e)**2/e for o,e in zip(df_table.values,Expected_value)])
chi_square_statestic=chi_square[0]+chi_square[1]
chi_square_statestic
```

```python
# from chi-square table(Î± = 0.05 and df=3)
critical_value=chi2.ppf(0.95,3)
critical_value
```

test statistic=1.515295 < critical_value=7.8147279\
failed to reject null hypothesis

ie,There is no dependency between male-female buyers rations and are similar across region.
