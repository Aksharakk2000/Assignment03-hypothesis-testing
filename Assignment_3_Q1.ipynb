**Q1.** A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
data = pd.read_csv("Cutlets.csv")
```

Null hypothesis(Ho):μ1 = μ2 ie,There is no difference in diameter of cutlets between two units
Alternate hypothesis(Ha): μ1 ≠ μ2 ie,There is significant diffrence in diameter of cutlets between two units

```python
data
```

```python
unit_A=pd.Series(data.iloc[:,0])
unit_A
```

```python
unit_B=pd.Series(data.iloc[:,1])
unit_B
```

Use 2-sample 2-tail ttest:

```python
#stats.ttest_ind(array1,array2)
p_value=stats.ttest_ind(unit_A,unit_B)
p_value
```

```python
p_value[1]
```

Here, p value is 0.472
comparing the p_value with level of significance at 5%(α=0.05),
 p_value is greater than level of significance so accept null hypothesis

ie, There is no difference in diameters of culters between two units
